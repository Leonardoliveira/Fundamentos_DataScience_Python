```python
#calculando a mediana
def mediana(x):
    arr = np.array(x) #Transforma lista em array
    arr.sort() #Reordenar o vetor
    if (arr.size % 2 == 0):#Localizar a posição da mediana
        A = (int(arr.size / 2)-1)
        B = int(arr.size/2)
        C = arr[A]
        D = arr[B]
        E = (C + D) / 2
        return (E)
    else:
        A = mt.ceil(arr.size/2)
        return (arr[A])
```

```python
mediana([1, 2, 3, 4, 5, 6])
```
