```python
#Fazendo uma função para calculo de valor presente liquido:
def npv_f(rate, cashflows):
    total = 0.0
    for i, cashflow in enumerate(cashflows):
        total += cashflow / (1 + rate) ** i
    print(total)    
    plt.plot(cashflows, marker = 'o') 
    return (total)
```

```python
#Calcular quantidade de parcelas, 50, vezes o valor da parcela, 1500, multiplicando pela taxa de 0.05 .
npv_f(0.05, [1500] * 50)
```

```python
#Fazendo um calculo para retornar o valor de parcela
def pmt_f(cashflows, rate):
    total = 0.0
    for i, cashflow in enumerate(cashflows):
        total += cashflow * ((1 + rate/100) ** i)
    plt.plot(cashflows, marker = 'o')
    return (total)
```

```python
pmt_f([1000, 2000, 3000], 5)
```

```python
#Utilizando operador while

cashflows = [550, -500, -500, -500, 1000]
r = 0.0
while (r < 1.0):
    r += 0.0001
    npv = npv_f(r, cashflows) 
    if(abs(npv) <= 0.001):
        print(r)
        
```

```python
#Função que valida notas e classifica de acordo com parametro de entrada.
def notas(nota):
    if nota > 100 or nota <0:
        print("Error!\nNota inválida!\n")
    else:
        if nota >= 90:
            print('A+')
        elif nota >= 85:
            print('A-')
        elif nota >= 75:
            print('B+')
        elif nota >= 65:
            print('B-')
        elif nota >= 55:
            print('C+')
        else:
            print('D-')
        
```

```python
#Função que valida presença
def presencaAluno(total, presenca):
    if presenca > total:
        print('Presença maior que o permitido.')
    else:
        if presenca < ((total * 75) / 100 ):
            print("Aluno Reprovado!")
        else:
            print("Aluno aprovado!")
```

```python
#Verificando nota e presença
sair = False
while sair == False:
    n = int(input("Digite a nota: "))
    notas(n)
    p = int(input("Presença do aluno: "))
    presencaAluno(15, p)
    s = int(input("Deseja sair?\n1. Sim\n2. Não\n"))
    if s == 1 :
        sair = True
print("Você saiu do sistema.")
```

Criando Anti Join com python

```python
import pandas as pd
```

```python
def anti_join(x, y, on):
    #Retorna linhas em x que não estão presentes em y
    ans = pd.merge(left=x, right=y, how='left', indicator=True, on=on)
    ans = ans.loc[ans._merge == 'left_only', :].drop(columns='_merge')
    return ans
    
```

```python
def anti_join_all_cols(x, y):
    #Retorna colunas em x que não estão presentes em y
    assert set(x.columns.values) == set(y.columns.values)
    return anti_join(x, y, x.columns.tolist())
```

Fazendo um gráfico 3D

```python
from sympy import symbols
from sympy.plotting import plot3d
x, y = symbols('x y')
```

```python
plot3d(x*y, (x, -5, 5), (y, -5, 5))
# Plot object containing:
# [0]: cartesian surface: x*y for x over (-5.0, 5.0) and y over (-5.0, 5.0)
```

```python
plot3d(x*y, -x*y, (x, -5, 5), (y, -5, 5))
# Plot object containing:
# [0]: cartesian surface: x*y for x over (-5.0, 5.0) and y over (-5.0, 5.0)
# [1]: cartesian surface: -x*y for x over (-5.0, 5.0) and y over (-5.0, 5.0)
```

```python
plot3d(x**4, -y+1, (x, -5, 5), (y, -5, 5))
```

```python
plot3d((x**2 + y**2, (x, -5, 5), (y, -5, 5)),
    (x*y, (x, -3, 3), (y, -3, 3)))
# Plot object containing:
# [0]: cartesian surface: x**2 + y**2 for x over (-5.0, 5.0) and y over (-5.0, 5.0)
# [1]: cartesian surface: x*y for x over (-3.0, 3.0) and y over (-3.0, 3.0)
```

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_surface
u, v = symbols('u v')
```

```python
plot3d_parametric_surface(cos(u + v), sin(u - v), u - v,
    (u, -5, 5), (v, -5, 5))
# Plot object containing:
# [0]: parametric cartesian surface: (cos(u + v), sin(u - v), u - v) for u over (-5.0, 5.0) and v over (-5.0, 5.0)
```
